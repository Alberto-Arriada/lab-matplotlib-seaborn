#### Use the appropriate plot to visualize the column `Gender`. There are 2 ways of doing it. Do it both ways.

*Hint: Use matplotlib and seaborn*

```python
# your code here- 1st way
Gender2=titanic.groupby('Gender')['PassengerId'].agg('count')
Gender2
```

```python
titanic['Gender'].hist()
```

```python
# your code here- 2nd way
fig1, ax1 = plt.subplots()
ax1.pie(Gender2, explode=(0,0.1), labels=['woman','man'], autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
plt.show()

```

#### Use the appropriate plot to visualize the column `Pclass` 


```python
# your code here
Pclass=titanic.groupby(['Pclass'])['PassengerId'].agg('count')
Pclass
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(Pclass, explode=(0.1,0.1,.1), labels=['Primera clase','Segunda clase','Tercera clase'], autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```
